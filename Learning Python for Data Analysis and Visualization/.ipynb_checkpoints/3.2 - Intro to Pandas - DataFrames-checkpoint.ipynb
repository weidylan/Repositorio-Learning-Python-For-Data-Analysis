```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import webbrowser # Grab a WebPage
```

```python
website = 'http://en.wikipedia.org/wiki/NFL_win-loss_records'
webbrowser.open(website) # Abre o site
```

```python
# Existem muitas maneiras de se criar um DataFrame

'''
Mas por agora, como estamos apenas interessados no funcionamento interno de um DataFrame e em como criar um a partir de zero.
Será mostrado como copiar e ler dados usando pandas.
'''

# Após copiar os dados de uma tabela (tabela do Wikipedia puxando dados da NFL) tudo aquilo fica na área de transferência.

# Com pandas podemos copiar e ler dados da área de transferência.

nfl_frame = pd.read_clipboard() # Lê tudo que está na área de transferência. 

# Pandas tem uma função embutida para tentar ler tudo o que está na sua área de transferência e configurálos para DataFrame.
```

```python
nfl_frame # Pandas transformou tudo que estava na área de transferência em um DataFrame.
```

```python
# Podemos pensar em um Data frame como quase uma planilha ou uma planilha do Excel.

# No topo temos os nomes das colunas e no canto da esquerda temos uma coluna com os índices do DataFrame.

# Essa coluna é criada automatiamente Pandas.

# Se você quiser pegar os nomes das colunas do DataFrameé bem simples.

nfl_frame.columns # Agora que temos os nomes das colunas podemos obter os dados dessas colunas.
```

```python
nfl_frame.index
```

```python
nfl_frame['Rank'] # Os dados retornados dessas coluna é uma Series.

# Retorna uma Serie com os valores da coluna.
```

```python
nfl_frame.Rank # Não podemos fazer esse método com colunas que possuem mais de um nome.

# Quando é assim pegamos o índice pelo nome
```

```python
nfl_frame['First NFL Season']
```

```python
# Pegando várias colunas

DataFrame(nfl_frame,columns=['Team', 'First NFL Season', 'Won']) 

# Criamos um novo DataFrame de dados do DataFrame raiz pegando algumas colunas.
```

```python
# O que vai acontece se colocarmos uma coluna que não existe?

DataFrame(nfl_frame,columns=['Team', 'First NFL Season', 'Won', 'Stadium']) 

# Essa coluna será preenchida com NaN = None.

# Temos um monte de valores Nulos mas o DataFrame ainda funciona.
```

```python
# Recuperando linhas do DataFrame

nfl_frame
```

```python
# Se você quiser recuperar apenas as primeiras linhas de um DataFrame

nfl_frame.head() # Retorna apenas 5 primeiras linhas
```

```python
nfl_frame.head(0) 
```

```python
nfl_frame.head(2) # Mas podemos especificar o número de linhas que será recuperada - Retorne as duas primeiras linhas
```

```python
# Para puxar as linhas finais utilizamos tail()

nfl_frame.tail() # O default é sempre 5
```

```python
nfl_frame.tail(2) # Me retorne as duas últimas linhas
```

```python
# Mas digamos que queremos recuperas linhas em um índice específico. Para que possamos resultar isso temos que fazer:

nfl_frame.ix[3] # Retorna uma Serie com tudo do índice 3.

# Agora temos um objeto retornado como todas as informações referentes a essa linha.
```

```python
# Uma indexação sempre começa pelo 0.

# Podemos também atribuir valores a colunas inteiras.

nfl_frame['Stadium'] = "Levi's Stadium"
```

```python
nfl_frame
```

```python
nfl_frame['Stadium'] = np.arange(5)
```

```python
nfl_frame
```

```python
# O que acontece se quisermos adicionar uma série a um DataFrame

stadiums = Series(["Levi's Stadium", "AT&T Stadium", "Mineirão", "Estádio Nacional", "Engenhão"], index=[4,0,2,1,3])
```

```python
stadiums
```

```python
nfl_frame['Stadium'] = stadiums
```

```python
nfl_frame
```

```python
# Também podemos excluir colunas inteiras.

del nfl_frame['Stadium']
```

```python
nfl_frame
```
